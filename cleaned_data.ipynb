### Merge with Part 2 (cleaned) and Research (new columns)

```python
part2 = pd.read_csv("clean_part2.csv")
research1 = pd.read_csv("hdb_absd_foreigner_2nd_loan.csv")
research2 = pd.read_csv("hdb_bto_mop.csv")
```

```python
part1 = data.iloc[ : , 0:40]
```

```python
cleaned_df = pd.merge(part1, part2, left_index=True, right_index=True, how='inner')
```

```python
research1.shape
```

```python
research2.shape
```

```python
r1_cols = ['Tranc_Year', 'absd_foreigner', 'ltv_limit_2nd_loan']
research1_df = research1[r1_cols]
```

```python
r2_cols = ['Tranc_Year','bto_launched', 'mop_flats']
research2_df = research2[r2_cols]
```

```python
research_df = pd.merge(research1_df, research2_df, on='Tranc_Year')
research_df.shape
```

```python
combined_df = pd.merge(cleaned_df, research_df, on='Tranc_Year', how="left")
```

```python
combined_df.columns
```

```python
combined_df.to_csv("combined.csv", index=False)
```
